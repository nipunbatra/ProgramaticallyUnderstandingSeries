### Problem statement

We have a biased coin (the bias being the probability to turn up heads). We decide with some prior information about the coin (before we observe any data).
Then, we toss the coin a few times and note down the observations. Based on the observations, we update our beliefs about the coin bias.
In this experiment, we will choose a coin with bias of 0.6. We wish to see if different priors and `enough data` can lead us to similar beliefs.

### Likelihood function

#### Conventions

* $\mu$ : Bias of the coin
* $N$ :  Number of coin tosses
* $m$ : Number of heads observed

Assume that we tossed the coin $N$ times, out of which it landed head $m$ times, then the likelihood of this is by: ${N\choose m}\mu^m(1-\mu)^{N-m}$


### Prior

The choice of priors has been a heavily debated topic. Everytime, we pick up the coin toss experiment, we usually see the Beta prior being talked about. 

Christopher Bishop in his famous Pattern Recognition and Machine Learning book says the following about the choice of Beta prior:
>Here we consider a form of prior distribution that has a simple
interpretation as well as some useful analytical properties. To motivate this prior,
we note that the likelihood function takes the form of the product of factors of the
form $\mu^x(1 − \mu)^{1−x}$ . If we choose a prior to be proportional to powers of $\mu$ and
(1 − $\mu$), then the posterior distribution, which is proportional to the product of the
prior and the likelihood function, will have the same functional form as the prior. This property is called conjugacy.

As we had seen before, the likelihood function has terms exponential in $\mu$ and $(1-\mu)$, we try to choose a prior such that it has similar form. Beta distribution has such a form.

Beta($\mu$|a, b)$\propto$$\mu^{a-1}(1 − \mu)^{b−1}$


### Posterior

Due the functional form of the chosen Beta prior, we get the Posterior$\propto$Prior.Likelihood as follows:

Posterior$\propto\mu^{m+a-1}(1-\mu)^{N-m+b-1}$

This can also be written as:

Beta($\mu$|m+a-1, N-m+b-1)

### Code setup

```python
# Customary imports
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
from matplotlib.animation import FuncAnimation
import seaborn as sns
np.random.seed(42)
%matplotlib inline

#Interactive widgets
from ipywidgets import StaticInteract, RangeWidget


# Stuff for animation
from tempfile import NamedTemporaryFile
VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=40, extra_args=['-vcodec', 'libx264'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)

from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))
```

### Beta distribution visualized

Say, we had a prior belief that out of 20 tosses, 10 would come out as heads. We would encapsulate it as Beta(10,10).  Now, say we had observed 20 heads and 5 tails in our dataset, we would obtain a posterior as Beta(10+20, 10+5). This suggests that the coin is more likely to give a heads. Let us play with an interactive widget and figure out how updating takes place based on observed evidence. In the widget below, I start with this prior of Beta(10,10) and by moving the slider for H and T, one can change the evidence which changes our belief. The plot shows the probability density function of $\mu$.

```python
def plot_beta(h,t):
    fig, ax = plt.subplots(figsize=(4,3))
    x = np.linspace(0,1,200)
    y = ss.beta.pdf(x, h, t)
    ax.plot(x,y)
    ax.set_xlim((0,1))
    ax.set_ylim((0,10))
    ax.set_title("#H: %d  #T: %d" %(h,t))
    return fig
StaticInteract(plot_beta, h=RangeWidget(10,30,10), t=RangeWidget(10,30,10))
```

### Animation!

Let us create an animation for the same. In this set of animations we vary the number of coin tosses, the priors and we see how our belief changes with our observations. The `real` $\mu$ for this particular coin is 0.6.

```python
#Class for updating distribution. Heavily influenced from Matplotlib's tutorial.I have added a prior 
#term as one of the additional arguments and also I have added the title to show the current count of heads and tails.
class UpdateDist(object):
    def __init__(self, ax, prior=[10,10],prob=0.5):
        self.prior =prior
        self.prob = prob
        self.line, = ax.plot([], [])
        self.x = np.linspace(0, 1, 200)
        self.ax = ax

        # Set up plot parameters
        self.ax.set_xlim(0, 1)
        self.ax.set_ylim(0, 20)
        self.ax.grid(True)

        # This vertical line represents the theoretical value, to
        # which the plotted distribution should converge.
        self.ax.axvline(prob, linestyle='--')

    def init(self):
        self.success = self.prior[0]
        self.total = self.prior[0] + self.prior[1]
        self.line.set_data([], [])
        return self.line,

    def __call__(self, i):
        # This way the plot can continuously run and we just keep
        # watching new realizations of the process
        if i == 0:
            return self.init()

        # Choose success based on exceed a threshold with a uniform pick
        if np.random.rand(1,) < self.prob:
            self.success += 1
        y = ss.beta.pdf(self.x, self.success + 1, (i + self.total - self.success) + 1,scale=1)
        self.line.set_data(self.x, y)
        self.ax.set_title("Prior: #H: %d #T %d \nObserved #H: %d  #T: %d" %(self.prior[0], self.prior[1], self.success, i + self.total - self.success+1) )
        return self.line,
```

#### Case I: Beta(10,10) prior, 100 observations

```python
fig = plt.figure(figsize =(6,3))
ax = fig.add_subplot(1, 1, 1)
plt.subplots_adjust(top=0.8)
ud = UpdateDist(ax, prior =[10,10],prob=0.6)
anim = FuncAnimation(fig, ud, frames=np.arange(100), init_func=ud.init,
        interval=100, blit=True)
display_animation(anim)
```

This prior seems reasonable and gets us close to our `true` bias.

#### Case II: Beta(20,0) prior, 100 observations

```python
fig = plt.figure(figsize =(6,3))
ax = fig.add_subplot(1, 1, 1)
plt.subplots_adjust(top=0.8)
ud = UpdateDist(ax, prior =[20,0],prob=0.6)
anim2 = FuncAnimation(fig, ud, frames=np.arange(100), init_func=ud.init,
        interval=100, blit=True)
display_animation(anim2)
```

A heavily `biased` prior of 20 heads and 0 tails as expected ends up a little to the right (more likely to show up heads).

#### Case III: Beta(0,20) prior, 100 observations

```python
fig = plt.figure(figsize =(6,3))
ax = fig.add_subplot(1, 1, 1)
plt.subplots_adjust(top=0.8)
ud = UpdateDist(ax, prior =[0,20],prob=0.6)
anim3 = FuncAnimation(fig, ud, frames=np.arange(100), init_func=ud.init,
        interval=10, blit=True)
display_animation(anim3)
```

As expected, a heavily `biased` prior favoring tails does not come close enough in 100 iterations. So what if we have more observations?

#### Case IV: Beta(0,20) prior, 600 observations

```python
fig = plt.figure(figsize =(6,3))
ax = fig.add_subplot(1, 1, 1)
plt.subplots_adjust(top=0.8)
ud = UpdateDist(ax, prior =[0,20],prob=0.6)
anim4 = FuncAnimation(fig, ud, frames=np.arange(600), init_func=ud.init,
        interval=10, blit=True)
display_animation(anim4)
```

We can thus conclude that given `enough` data, the choice of priors should be immaterial. Also, when `enough` data is available, this approach would give a similar result to the Maximum Likelihood approach.
